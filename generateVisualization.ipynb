```python
import pandas as pd
import copy
# Get model results from ResultingModelAccuracies.csv
df_model_accuracies_info = pd.read_csv("ResultingModelAccuracies.csv")
df_model_accuracies_info.head(25+3)
```

```python
df_VADER_info = copy.deepcopy(df_model_accuracies_info[df_model_accuracies_info["Model"] == "VADER"])
df_VADER_info
```

```python
df_TextBlob_info = copy.deepcopy(df_model_accuracies_info[df_model_accuracies_info["Model"] == "TextBlob"])
df_TextBlob_info.reset_index(inplace = True, drop = True)
df_TextBlob_info
```

```python
df_FinBERT_info = copy.deepcopy(df_model_accuracies_info[df_model_accuracies_info["Model"] == "FinBERT"])
df_FinBERT_info.reset_index(inplace = True, drop = True)
df_FinBERT_info
```

```python
df_LSTM_info = copy.deepcopy(df_model_accuracies_info[df_model_accuracies_info["Model"] == "LSTM"])
df_LSTM_info.reset_index(inplace = True, drop = True)
df_LSTM_info
# df_LSTM_info["Model"] = pd.to_numeric(df_LSTM_info["Model"])
```

```python
LSTM_info = df_LSTM_info.to_numpy().tolist()
LSTM_info = sorted(LSTM_info, key=lambda x: x[3], reverse=False)
LSTM_info
```

```python
df_sorted_LSTM_info = pd.DataFrame(LSTM_info, columns=["Model","Lowest Accuracy","Highest Accuracy","Average Accuracy"])
df_sorted_LSTM_info
```

```python
cumulative_LSTM_info = [["LSTM", df_LSTM_info["Lowest Accuracy"].min(), df_LSTM_info["Highest Accuracy"].max(), df_LSTM_info["Average Accuracy"].median()]]
df_cumulative_LSTM_info = pd.DataFrame(cumulative_LSTM_info, columns=["Model","Lowest Accuracy","Highest Accuracy","Average Accuracy"])
df_cumulative_LSTM_info
```

```python
df_cumulative_model_accuracies_info = pd.concat([df_VADER_info, df_TextBlob_info, df_cumulative_LSTM_info, df_FinBERT_info])
df_cumulative_model_accuracies_info.reset_index(inplace = True, drop = True)
df_cumulative_model_accuracies_info
```
